# Merging Goodreads to NYT data

```python
import pandas as pd
```

# Importing NYT Bestseller data

```python
books = pd.read_csv('books.csv')
```

```python
books
```

```python
books.rename(columns={'title': 'title_query'}, inplace=True)
```

```python
books.head()
```

```python
books.shape
```

# Importing Goodreads Data

```python
goodreads = pd.read_csv('goodreads_data.csv')
```

```python
goodreads
```

```python
df = pd.merge(books, goodreads, how='left', on='title_query')
```

```python
df.head()
```

```python
df.shape
```

# Cleaning out some of the duplicates do to index numbers

```python
data = df[['title_query','author_x','author_y','min_rank', 'max_weeks_on_list','author_id', 'average_rating', 'book_id', 'ratings_count', 'title', 'title_match', 'gender', 'num_pages']].drop_duplicates()
```

```python
data.head()
```

```python
data.shape
```

```python
data = df[['title_query','author_x','author_y','min_rank', 'max_weeks_on_list','author_id', 'average_rating', 'book_id', 'ratings_count', 'title','title_match', 'gender', 'num_pages']].drop_duplicates()
```

```python
data.shape
```

# Rename columns so we know which were the NYT authors and which were the goodreads authors

```python
data.rename(columns={'author_x': 'author_nyt'}, inplace=True)
```

```python
data.rename(columns={'author_y': 'author_gr'}, inplace=True)
```
